### Claire 15 Plot for Paper

```python
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.nonparametric.smoothers_lowess import lowess
import seaborn as sns
```

```python
claire_raw = pd.read_csv('CSV/wheat/Claire_raw.csv')
claire_summary = pd.read_csv('CSV/wheat/Claire_summary.csv')
```

```python
claire15_raw
```

```python
claire15_raw = claire_raw[claire_raw['Name'] == 'claire_0015']
claire15_summary = claire_summary[claire_summary['Name'] == 'claire_0015']
```

```python
claire15_summary
```

```python
claire15_raw['Avg RHL'] = claire15_raw.loc[:, ['RHL 1', 'RHL 2']].mean(axis=1)
```

```python
claire15_raw
```

```python
claire15_raw[claire15_raw['Distance From Root Tip (mm)'] >= 1.96]
```

```python
line1 = lowess(claire15_raw['RHL 1'], claire15_raw['Distance From Root Tip (mm)'], frac=0.15)
line2 = lowess(claire15_raw['RHL 2'], claire15_raw['Distance From Root Tip (mm)'], frac=0.15)
avg = lowess(claire15_raw['Avg RHL'], claire15_raw['Distance From Root Tip (mm)'], frac=0.15)
```

```python

sns.set_style('whitegrid')
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 20
plt.rcParams['figure.figsize'] = (30,12)
```

```python
fig, ax = plt.subplots()
plt.scatter(x=claire15_raw['Distance From Root Tip (mm)'],
            y=claire15_raw['RHL 1'],
            color='darkmagenta',
            marker='*',
            s=200,
            alpha=0.3,
            )
plt.scatter(x=claire15_raw['Distance From Root Tip (mm)'],
            y=claire15_raw['RHL 2'],
            color='lightseagreen',
            marker='^',
            alpha=0.3,
            s=200
            )
plt.plot(line1[:,0], line1[:,1], linewidth=7, linestyle='dashed', color='darkmagenta')
plt.plot(line2[:,0], line2[:,1], linewidth=7, linestyle='dashdot', color='lightseagreen')
# elongation zone
plt.plot((claire15_summary['Elongation Zone Start (mm)'], claire15_summary['Elongation Zone Start (mm)']), (0, 5), linewidth=5, linestyle='dashed', color='firebrick')
plt.plot((claire15_summary['Elongation Zone Stop (mm)'], claire15_summary['Elongation Zone Stop (mm)']), (0, 5),linewidth=5, linestyle='dashed', color='firebrick')
plt.annotate(text='', xy=(4,2.8), xytext=(8,2.8), arrowprops=dict(arrowstyle='<->', shrinkA=0, shrinkB=0, linewidth=3, color='firebrick', mutation_scale=40))
plt.annotate(text='Elongation Zone', xy=(4.6,2.9), fontsize=22, color='firebrick')
plt.annotate(text='9.02 mm', xy=(5.2, 2.65), fontsize=22, color='firebrick')
plt.annotate(text='1.96 mm', xy=(0.6, 2.6), fontsize=18, color='firebrick')
plt.annotate(text='10.98 mm', xy=(11.2, 2.6), fontsize=18, color='firebrick')
plt.annotate(text='', xy=(11,1.72), xytext=(13,2.5), arrowprops=dict(arrowstyle='->', shrinkA=0, shrinkB=0, linewidth=3, color='red', mutation_scale=20))
plt.annotate(text='1.72 mm', xy=(13.1, 2.5), color='red', fontsize=18 )
plt.annotate(text='', xy=(1.96, 0), xytext=(3.96,0.25), arrowprops=dict(arrowstyle='->', shrinkA=0, shrinkB=0, linewidth=3, color='red', mutation_scale=20))
plt.annotate(text='0 mm', xy=(4.05, 0.25), color='red', fontsize=18 )
plt.annotate(text=r'Gradient = $\frac{1.72 - 0}{10.98 - 1.96} = 0.19$', xy=(6,0.5), color='firebrick')
# max rhl delta
plt.annotate(text='', xy=(claire15_summary['Max RHL Segment Pos (mm)'], 1.7), xytext=(claire15_summary['Max RHL Segment Pos (mm)'],2.3), arrowprops=dict(arrowstyle='<->', shrinkA=0, shrinkB=0, linewidth=2, color='darkblue', mutation_scale=20))
plt.scatter(15.4901,1.667, color='darkmagenta', marker='*',s=400)
plt.scatter(15.4901,2.3235, color='lightseagreen', marker='^', s=400)
plt.plot((13,15.490),(1.2,2), color='darkblue', linewidth=2)
plt.annotate(text=r'Largest RHL $\delta$: 0.66 mm at 15.5 mm along root', xy=(12, 1.1), color='darkblue')
# avg rhl
plt.plot((0,25 ), (claire15_summary['Avg RHL (mm)'], claire15_summary['Avg RHL (mm)']), linewidth=3, linestyle='dashdot', color='darkorange')
plt.annotate(text=r'Average RHL: 1.45 mm', xy=(2.5, 1.5), fontsize=22, color='darkorange')

plt.ylim(0, max(claire15_raw['RHL 1']) * 1.5)
plt.xlim(0, 25)
plt.xlabel('Distance From Root Tip (mm)')
plt.ylabel('Root Hair Length (mm)')
plt.legend(loc='upper right')
for _, spine in ax.spines.items():
        spine.set_visible(True)
        spine.set_color('black')
        spine.set_linewidth(2)

```
